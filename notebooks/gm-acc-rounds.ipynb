```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
headers = ['expID','nSites', 'lstm_cells', 'threshold', 'batch_size', 'max_iters', 'accuracy', 'rounds']
df = pd.read_csv('../simulation/exps.csv', usecols=headers)
```

#### LSTM Cells in respect of accuracy and rounds ####

```python
sub_df = df[df.expID == 0]
_sorted = sub_df.sort_values('lstm_cells')
x = _sorted['lstm_cells']
y = _sorted['accuracy']

plt.plot(x,y)
plt.legend(['Accuracy'], loc='upper left')
plt.grid(True)
plt.show()

plt.plot(x,y)
plt.legend(['Rounds'], loc='upper left')
plt.grid(True)
plt.show()
```

#### Threshold in respect of accuracy and rounds ####

```python
sub_df = df[df.expID == 0]
_sorted = sub_df.sort_values('threshold')
x = _sorted['threshold']
y = _sorted['accuracy']

plt.plot(x,y)
plt.legend(['Accuracy'], loc='upper left')
plt.grid(True)
plt.show()

y = _sorted['rounds']
plt.plot(x,y)
plt.legend(['Rounds'], loc='upper left')
plt.grid(True)
plt.show()
```

#### Optimizer max iterations in respect of accuracy and rounds ####

```python
sub_df = df[df.expID == 0]
_sorted = sub_df.sort_values('max_iters')
x = _sorted['max_iters']
y = _sorted['accuracy']

plt.plot(x,y)
plt.legend(['Accuracy'], loc='upper left')
plt.grid(True)
plt.show()

y = _sorted['rounds']
plt.plot(x,y)
plt.legend(['Rounds'], loc='upper left')
plt.grid(True)
plt.show()
```
